```python
def add_country(dictionary, region_id):
    query = "INSERT INTO countries(country_id, country_name, region_id) VALUES (%s, %s, %s)"
    
    values = []
    for country_id, country_name in dictionary.items():
        values.append((country_id, country_name, region_id))
    
    mycursor.executemany(query, values)
    cnx.commit()
```

```python
# The function wants to insert songs into an SQL table named songs
def add_songs(dictionary, decade_id):
    songs_query = "INSERT INTO songs(song_title, genre, popularity) VALUES(%s, %s, %s)"
    relation_query = "INSERT INTO decade_of_song(decade_id, song_id) VALUES (%s, %s)"

    for i in range(len(dictionary['nombre'])):
        #Insert song
        val_song = (
        str(dictionary['nombre'][i]).strip("[]'"),
        str(dictionary['genero'][i]),
        int(dictionary['popularity'][i])
)

        mycursor.execute(songs_query, val_song)
        cnx.commit()

        # Get last inserted song ID
        song_id = mycursor.lastrowid

        # Link song with its period (lustro)
        val_relation = (decade_id, song_id)
        mycursor.execute(relation_query, val_relation)
        cnx.commit()

'''What this code does:

            - Inserts each song with its title, genre, and popularity into the songs table.

            - Automatically obtains the song_id generated by MySQL.

            - Inserts a relationship between that song_id and the corresponding decade_id in the song_decade table.'''
```
