```python
## - Explicar como usar la libreria spotipy 
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
# Coloca tus credenciales aquí
CLIENT_ID = input ('inserta tu CLIENT ID')
CLIENT_SECRET = input ('inserta tu CLIENT_SECRET')

# Autenticación
auth_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
sp.search(q = "genre:punk year:2000", limit=10, offset=0, type='track,album', market="ES")
```

```python
datos_2000 = sp.search(q = "genre:punk year:2000", limit=10, offset=0, type='track,album', market="ES")
```
