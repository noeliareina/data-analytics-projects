```python
limit = 50
popularidad = []
for page in range(limit, limit+2):
    datos_2024 = sp.search(q = "year:2024", limit=50, offset={page}, type='track', market="ES")
    for element in datos_2024['tracks']['items']:
        prueba['popularity'].append(element['popularity'])
        prueba['cancion'].append(element['name'])
        artistas = []
        for artista in element['artists']:
            artistas.append(artista['name'])
        prueba['artista'].append(artistas)
    
```

```python
popularidad.sort(reverse=True)
print(popularidad) # Esta es la popularidad de las canciones. Nos van a servir para ordenar las canciones por su popularidad (?)
```
