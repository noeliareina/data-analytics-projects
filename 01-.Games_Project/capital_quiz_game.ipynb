```python
# --- Welcome message and player setup ---
print("üåç Welcome to the Country & Capital Quiz!")
# First of all, could you please give us your name ?
player_name = input("Enter your nickname: ").strip().capitalize()
print(f"\nHi {player_name}! Let's see how many capitals you can guess correctly.\n")

# --- Data dictionary: capitals and countries ---
dict_geo_cap = {
    "Serbia": "Belgrade", "Canada": "Ottawa", "Colombia": "Bogota",
    "France": "Paris", "England": "London", "Brazil": "Brasilia",
    "Bulgaria": "Sofia", "Costa Rica": "San Jose", "Algeria": "Algiers",
    "Finland": "Helsinki", "Georgia": "Tbilisi", "Japan": "Tokyo",
    "Bolivia": "La Paz", "Portugal": "Lisbon", "Nepal": "Kathmandu",
    "Germany": "Berlin", "Italy": "Rome", "Spain": "Madrid",
    "Mexico": "Mexico City", "Australia": "Canberra",
    "China": "Beijing", "Russia": "Moscow", "Greece": "Athens",
    "Egypt": "Cairo", "Norway": "Oslo", "Sweden": "Stockholm",
    "Switzerland": "Bern", "Poland": "Warsaw", "India": "New Delhi",
    "Argentina": "Buenos Aires", "Chile": "Santiago", "Peru": "Lima",
    "Venezuela": "Caracas", "Uruguay": "Montevideo", "Paraguay": "Asuncion",
    "Ecuador": "Quito", "Panama": "Panama City", "Cuba": "Havana",
    "Honduras": "Tegucigalpa", "Nicaragua": "Managua", "South Africa": "Pretoria",
    "South Korea": "Seoul", "North Korea": "Pyongyang", "Turkey": "Ankara",
    "Ireland": "Dublin", "Netherlands": "Amsterdam", "Belgium": "Brussels",
    "Denmark": "Copenhagen", "Ukraine": "Kyiv", "Iceland": "Reykjavik",
    "Hungary": "Budapest"
}

# Applying .capitalize() to keys and values:
'''dict_geo_cap = {k.capitalize(): v.capitalize() for k, v in dict_geo.items()}'''

# Verifying the firsts pairs key-values:
'''list(dict_geo_cap.items())[:10]'''

import random
# Variables to track the score, starting from zero
score = 0
total_questions = 5

# Randomly select 5 questions, we need to transform the dictionary in a list to use random.sample (5 pairs)
questions = random.sample(list(dict_geo_cap.items()), total_questions)

# Main quiz loop
for country, capital in questions:
    answer = input(f"Which capital has {country} as its country? ").strip().title()
    if answer == capital:
        print("‚úÖ Correct!")
        score += 1
    else:
        print(f"‚ùå Wrong. The correct answer is {capital}.")
    print("-" * 30)

# Final score and feedback 
print(f"\n{player_name}, your final score is {score}/{total_questions}")

if score == total_questions:
    print("üèÜ Perfect! You‚Äôre a geography master!")
elif score >= 3:
    print("üëè Good job! You know the capitals quite well.")
else:
    print("üí¢ Keep practicing! You‚Äôll get better next time.")

```
