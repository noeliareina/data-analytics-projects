```python
# We import the ‚Äòrandom‚Äô library so that the programme responds randomly, allowing us to play against it.
import random

# --- Game Setup ---
print("Welcome to Rock, Paper, Scissors!")

player_name = input ("Please enter your nickname: ").capitalize()
print(player_name)

# We create a variable that marks the increase in score from 0.
player_score = 0
computer_score = 0

while player_score < 3 and computer_score < 3:
    player_choice = input("Choose: rock, paper, or scissorsüóøüßª‚úÇÔ∏è").lower() 
    computer_choice = random.choice(["rock","paper","scissors"])  
    print(player_choice, "vs", computer_choice)


    if player_choice not in ["rock", "paper", "scissors"]:
        print("Invalid option. You must choose between rock, paper, or scissors.üò°")   
        continue 


    if player_choice == computer_choice:
            print("It's a draw! ü§ù")
    elif player_choice == 'rock' and computer_choice == 'scissors':
            print('You have earned a point! ‚úîÔ∏è')
            player_score += 1
    elif player_choice == 'paper' and computer_choice == 'rock':
            print('You have earned a point! ‚úîÔ∏è')
            player_score += 1
    elif player_choice == 'scissors' and computer_choice == 'paper':
            print('You have earned a point! ‚úîÔ∏è')
            player_score += 1
    else:
            print('One point off, loser! ü§™')
            computer_score += 1
        
       

print ("---------------------")

if player_score == 3:
    print(f"Your score is 3, you win {player_name}.ü•≥")
    
else:
    print ("The computer program has 3 points, you loss.üò≠")




```

```python
# We import the ‚Äòrandom‚Äô library so that the programme responds randomly, allowing us to play against it.
import random

# --- Game Setup ---
print("Welcome to Rock, Paper, Scissors!")

player_name = input ("Please enter your nickname: ").capitalize()
print(player_name)
# We need to create a list for the game 3 options.
options = ["rock", "paper", "scissors"]

# We define the function play_game
def play_game():
       print(f"\nHi {player_name}! Let's play üëá")
       player_choice = input("Choose rock, paper, or scissors: ").lower()
       computer_choice = random.choice(options)
       
       print(f"\n{player_name} chose: {player_choice}")
       print(f"The computer chose: {computer_choice}")

       if player_choice == computer_choice:
          print("It's a draw! ü§ù")
       elif (player_choice == "rock" and computer_choice == "scissors") or \
         (player_choice == "paper" and computer_choice == "rock") or \
         (player_choice == "scissors" and computer_choice == "paper"):
           print("You win! üéâ")
       else:
           print("You lose üò¢")

# --- Game Loop ---
while True:
    play_game()
    play_again = input("\nDo you want to play again? (yes/no): ").lower()
    if play_again != "yes":
        print(f"\nThanks for playing, {player_name}! üëã")
        break
```
